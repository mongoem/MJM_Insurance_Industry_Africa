```python
import pandas as pd
```

```python
# Load the data
insurance_data_cleaned = pd.read_excel('insurance_claims_cleaned.xlsx')

# Display the first few rows
insurance_data_cleaned.head()
```

```python
import matplotlib.pyplot as plt

# Select only numerical columns for histogram plotting
numerical_data = insurance_data_cleaned.select_dtypes(include=['float64', 'int64'])

# Plotting histograms for all numerical features
numerical_data.hist(bins=15, figsize=(15, 10), layout=(5, 4))
plt.tight_layout()  # Adjusts subplot parameters to give a nicer layout
plt.show()
```

```python
bin_edges = [0, 30, 55, 100]  # Define the bin edges
bin_labels = ['Young Adult', 'Middle Aged', 'Elderly']  # Corresponding labels for each bin

# Create a new column based on the bin labels
insurance_data_cleaned['ages_category'] = pd.cut(insurance_data_cleaned['age'], bins=bin_edges, labels=bin_labels)
```

```python
insurance_data_cleaned.head()
```

```python
bin_edges_customer = [0, 25, 150, 500]  # Define the bin edges
bin_labels_customer = ['New Client', 'Established Client', 'Long-Term Client']  # Corresponding labels for each bin

# Create a new column based on the bin labels
insurance_data_cleaned['customer_category'] = pd.cut(insurance_data_cleaned['months_as_customer'], bins=bin_edges_customer, labels=bin_labels_customer)
```

```python
insurance_data_cleaned.head()
```

```python
insurance_data_cleaned["Contract Years"] = insurance_data_cleaned["months_as_customer"]/12
```

```python
insurance_data_cleaned.head()
```

```python
insurance_data_cleaned['total_premiums_paid'] = (insurance_data_cleaned['policy_annual_premium']/12) * insurance_data_cleaned['months_as_customer']
```

```python
insurance_data_cleaned['net_value_of_customer'] = insurance_data_cleaned['total_premiums_paid'] - insurance_data_cleaned['total_claim_amount']
```
